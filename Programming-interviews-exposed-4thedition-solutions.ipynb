#### PROBLEM 1: 
Write a program that prints the numbers from 1 to 100. If a number is divisible by 3, print “Fizz” instead of the number, if a number is divisible by 5, print “Buzz” instead of the number, and if a number is divisible by both 3 and 5, print “FizzBuzz” instead of the number.

```python
# Simple quick solution
import math # you need math.fmod as python % mod operator works differently.

def fizzBuzz(minRange, maxRange):
    
    # Loop over through the provided range of numbers and check for divisibility//
    for iter in range(minRange, maxRange+1): #maxRange+1 cause python range function does include the right limit's value//
        divisibleThree = int(math.fmod(iter, 3))
        divisibleFive = int(math.fmod(iter, 5))
        
        if (divisibleThree == 0) & (divisibleFive == 0):
            print("FizzBuzz")
        elif divisibleThree == 0:
            print("Fizz")
        elif divisibleFive == 0:
            print("Buzz")
        else:
            print(iter)
            
            
            
if __name__ == "__main__":
    minRange = 1
    maxRange = 100 # always good to ask the interviewer as to how they want you to read inputs and display outputs//
    
    fizzBuzz(minRange, maxRange)
```

```python
# recursive approach for FizzBuzz problem//

def fizzBuzzRecursive(minRange, maxRange):
    divisibleThree = int(math.fmod(minRange, 3))
    divisibleFive = int(math.fmod(minRange, 5))
    
    if (divisibleThree == 0) & (divisibleFive == 0):
        print("FizzBuzz")
    elif divisibleThree == 0:
        print("Fizz")
    elif divisibleFive == 0:
        print("Buzz")
    else:
        print(minRange)
    
    if minRange < maxRange:
        fizzBuzzRecursive(minRange+1, maxRange)
    elif minRange == maxRange:
        return 
    
if __name__ == "__main__":
    minRange = 1
    maxRange = 100 # always good to ask the interviewer as to how they want you to read inputs and display outputs//
    
    fizzBuzzRecursive(minRange, maxRange)
```

#### PROBLEM 2:
Write a function that reverses a string without using any library
functions.

```python
def reverseString(inputString):
    reversedString = []
    stringLength = len(inputString)
    
    for iter in range(stringLength-1, -1, -1):
        reversedString.append(inputString[iter])
        
    return ''.join(reversedString)

if __name__ == "__main__":
    print(reverseString('Hi, there! /n How are you doing?'))
    print(reverseString('savitaa venkateswaran'))
```

#### PROBLEM 3:
Given an unsorted list of integers, write a function that returns a
new list with all duplicate values removed.

```python
# simple solution convert list to set//
if __name__ == "__main__":
    arr = [4, 4, 1, -1, 5, 6, 5, 6, -1, 7, 100, -290, 4, 1, -2, -450, 1000, -450] #unsorted array of integers//
    
    print(set(arr)) #order not preserved
```

```python
# If original order needs to be preserved, keeping/retaining the first occurence of the integer.

def removeDuplicates(arr):
    recordOccurences = {} 
    arrayLength = len(arr)
    count = 0
    
    while count <= arrayLength-1:
        value = arr[count]
        if value in recordOccurences:
            arr.pop(count)
            arrayLength -= 1
        else:
            recordOccurences.__setitem__(value, 1)
            count += 1
            
    return arr

if __name__ == "__main__":
    arr = [4, 4, 1, -1, 5, 6, 5, 6, -1, 7, 100, -290, 4, 1, -2, -450, 1000, -450] #unsorted array of integers//

    print(removeDuplicates(arr)) #order not preserved
```
