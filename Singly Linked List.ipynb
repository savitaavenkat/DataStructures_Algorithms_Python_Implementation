```python
# Author: Savitaa Venkateswaran

# Create every node as an object//
class Node:
    # Initialize a node with data/value and its next pointer to be NULL//
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
# Create a linkedlist object which is a collection of nodes linked together//
class LinkedList:
    # Initialize the head and tail nodes of the linked list to intially be NULL//
    def __init__(self):
        self.head = None
        self.tail = None
    
    # Adds new node to the end of linkedlist//
    def addNode(self, node): # O(1) operation
        if self.head != None:
            self.tail.next = node
            self.tail = self.tail.next
        else:
            self.head = node
            self.tail = self.head
            
    def addNodeBeginning(self, node): # O(1) operation
        if self.head != None:
            node.next = self.head
            self.head = node
        else:
            self.head = node
            self.tail = self.head
        
    def deleteNode(self): # O(1) operation
        # deletes last node of the linkedlist
        if self.head != None:
            node = self.head
            if node.next == None:
                self.tail = None
                self.head = self.tail
            else:  
                while node.next.next != None:
                    node = node.next     
                self.tail = node 
                self.tail.next = None
        else:
            pass
        
    def deleteNthNode(self, n): # O(N) operation
        # deleted node at the nth position
        node = self.head
        
        # Check if the list is not empty, head != NULL//
        if node != None:
            # Get length of list
            listLength = int(self.findLength())
            # If the node to be deleted is the last element call deleteNode()//
            if n == listLength:
                self.deleteNode()
                return "Node deleted."
            # Else if node to be deleted is the head of linked list (first element, simple remove it)//
            elif n == 1:
                self.head = node.next
                self.head.next = node.next.next
                node.next = None
                return "Node deleted."  
            # Else delete nth node//
            else:
                count = 1
                #parse through the linkedlist n-1 times//
                for iter in range(1, n):
                    if node.next.next != None:
                        prev = node
                        node = node.next
                        next = node.next
                        count += 1        
                #print(prev.data, node.data, next.data, count)
                # If such an nth element exists, delete it
                if count == n:
                    prev.next = next
                    node.next = None
                    return "Node deleted."
                # Else if n is greater than linkedlist length, do nothing//
                else:
                    "No node to be deleted."
        else:
            return "List empty, cannot perform delete on an empty list."
        
    def printList(self): # O(N) operation
        # Get all nodes in the
        node = self.head
        linklist = []
        while node != None:
            linklist.append(str(node.data)+str("->"))
            node = node.next
        return linklist
    
    def findLength(self): # O(N) operation
        node = self.head
        count = 0
        while node != None:
            node = node.next
            count += 1
        return str(count)
        
```
